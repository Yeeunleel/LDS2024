## 1. 단어구름 그리기

우리는 [숙제05] 및 10강 실습 코드에서 이미 단어구름(word cloud)을 그려본 적이 있다.

기본적으로는 `wordcloud` 모듈을 설치하고 `WordCloud` 객체를 초기화한 뒤, `generate_from_frequencies(Counter)` 메소드로 그림을 생성하면 된다.

이 과정은 10강 실습 코드에서 이미 `plot(Counter)` 함수로 정의해 보았다.



````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

palettes = ['spring', 'summer']
font_path = '../../fonts/NotoSansCJKkr-Regular.otf'

def plot(counts):
  for label in counts.keys():
    plt.subplot(1, len(counts), label+1) # 작은 그림: (1, 2) 모양의 그림
    wc = WordCloud(font_path=font_path, width=600, height=600, colormap=palettes[label])
    gen = wc.generate_from_frequencies(counts[label]) # 인자: Counter
    plt.axis('off')
    plt.imshow(gen) # 작은 그림
  
  plt.show() # 큰 그림 출력
````

기존의 카운터로 단어구름을 일단 그려 보자.

````python
plot(counts)
````

# 2. 단어구름 개선하기 (필수, 4점)

### 기존 카운터로 그린 단어구름의 문제점과 해결책

방금 그린 단어구름에서는 왼쪽의 부정적인 영화평과 오른쪽의 긍정적인 영화평에서 공통적으로 많이 나타나는 단어를 찾을 수 있다.

위에서 출력해 본 상위 10개 단어만 하더라도 영화, 점, 진짜, 정말, 이, 것 6개가 겹쳤다.

이렇게 겹치는 단어가 많으면, 부정적인 영화평과 긍정적인 영화평 사이의 차이를 명확하게 살펴보기 어렵다.

이 문제를 해결하기 위해 아래 셀에서 새로운 카운터 `counts_exclusive`를 만들라.

````python
from copy import deepcopy
# counts_exclusive = counts # NEVER DO THIS
# counts_exclusive = counts.copy() # DON'T DO THIS, AT LEAST FOR THIS HOMEWORK
counts_exclusive = deepcopy(counts) # DO THIS

# DO SOMETHING HERE 
# 각 영화평에서 자주 사용된 단어를 상위 30개씩 추출하고, 그중 겹치는 단어들을 단어구름에서 제외한다.  

# 부정적 영화평에서 가장 자주 사용된 키워드 30개
neg_common = set([word for word, _ in counts[0].most_common(30)])

# 긍정적 영화평에서 가장 자주 사용된 키워드 30개
pos_common = set([word for word, _ in counts[1].most_common(30)])

# 부정적/긍정적 영화평에서 모두 등장한 단어 찾기
common_words = neg_common & pos_common

# 부정적, 긍정적 카운터에서 공통 단어 제거
for word in common_words:
    if word in counts_exclusive[0]:
        del counts_exclusive[0][word]
    if word in counts_exclusive[1]:
        del counts_exclusive[1][word]
````

````python
common_words
````

이제 새로 만든 카운터에서 부정적인 영화평의 상위 10개 단어를 뽑아 보자.

단, 출력 결과가 예시와 꼭 일치할 필요는 없다.

````python
# Run this cell again
print(counts_exclusive[0].most_common(10))
````

다음으로 긍정적인 영화평의 상위 10개 단어를 뽑아 보자. 두 목록이 서로 겹치지 않는다면 잘 된 것이다.

````python
# Run this cell again
print(counts_exclusive[1].most_common(10))
````

새로 만든 카운터로 단어구름을 그렸을 때 두 부분집합의 차이가 한눈에 드러나면 된다.

그림이 예시와 꼭 일치할 필요는 없다.


````python
# Run this cell again
plot(counts_exclusive)
````

## 3. [숙제09심화] 보너스 문제 (선택, +1점)

이 문제는 필수적으로 답할 필요는 없다.

앞서 `counts_exclusive`를 설정할 때 `counts`의 값을 받아오기 위해 `copy.deepcopy()` 함수를 사용했다.

```
from copy import deepcopy
# counts_exclusive = counts # NEVER DO THIS
# counts_exclusive = counts.copy() # DON'T DO THIS, AT LEAST FOR THIS HOMEWORK
counts_exclusive = deepcopy(counts) # DO THIS
```

이 과제에서 `counts`의 값을 복사할 때 단순 할당문이나 `dict.copy()` 메소드를 쓰면 안 되는 이유를 아래 셀에서 설명하라.

답: 단순 할당을 할 경우 counts_exclusive와 counts는 동일한 메모리 주소를 가리키게 된다. 따라서 counts_exclusive에서 어떤 값을 수정하면 원본인 counts도 함께 수정된다.

dict.copy()는 얕은 복사를 수행한다. 이 때 딕셔너리 자체는 새로운 메모리 공간에 할당되지만, 그 내부에 포함된 가변 객체(리스트, 딕셔너리 등)는 같은 참조를 가진다. 즉, 내부의 가변 객체는 공유되기 때문에, 이를 수정하면 원본의 해당 부분도 함께 변경된다. 따라서 원본과 완전히 독립된 복사본을 생성하기 위해서는 deepcopy()를 사용해야 한다.
